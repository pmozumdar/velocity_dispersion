```python
from veldis.veldis import Veldis
```

```python
spec = Veldis("../galaxy_spectra/lris-longslit-red-1120/spec1d_r201121_0043-J0805+6919_LRISr_2020Nov21T152211.856.fits",
              informat='pypeit')
```

```python
%matplotlib notebook
spec.plot()
```

```python
spec.smooth(11)
```

```python
spec.cal_redshift(Mgb=6302.52)
```

```python
spec.smooth(7)
spec.mark_lines('abs', z=0.2176, usesmooth=True)
```

```python
spec.trimspec(wavrange=[5500, 8820])
```

```python
mask_reg = [(8.8348, 8.8450), (8.9338, 8.9521), (9.0142, 9.0172)]
```

```python
spec.cal_parm(z=0.2176, norm=False)
```

```python
temp_spec = spec.gen_rebinned_temp(lib_path='../TEXT/*', sig_ins=1.68)
```

```python
spec.cal_veldis(temp_spec=temp_spec, mask_reg=mask_reg, degree=[4, 21], show_weight=True)
```

```python
spec.plot_veldis(ylim=[300,355])
```
