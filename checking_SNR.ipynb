```python
lens_sys = [B0445_spec, B0631_spec, J0147_spec, J0818_spec, J0924_spec, J1433_spec, J1817_spec]
lens_name = np.array(['B0445+123', 'B0631+519', 'J0147+4630', 'J0818-2613', 'J0924+0219',
                      'J1433+6007', 'J1817+2729'])
```

```python
for i,sp in enumerate(lens_sys):
    fg = plt.figure(figsize=(10, 4))
    sp.smooth(7, fig=fg)
    sp.mark_lines('abs', z=lens_z[i], usesmooth=True, markatm=False)
```

### Lets calculate SNR using the flux with the restframe wavelength around Gband and Fe4383

```python
rest_wav = np.array([4265.00, 4395.00])# Gband is marked at 4305 and Fe4383 at 4383
median_snr = []
mean_snr = []
```

```python
for i, sp in enumerate(lens_sys):
    index = sp.closest_wavrange([rest_wav*(1+lens_z[i])],verbose=False)
    ssp = spec1d.Spec1d(wav=sp['wav'][index[1][0][0]:index[1][0][1]+1], 
                        flux=sp['flux'][index[1][0][0]:index[1][0][1]+1],
                        var=sp['var'][index[1][0][0]:index[1][0][1]+1])
    fg = plt.figure(figsize=(6, 3))
    ssp.smooth(1, fig=fg)
    ssp.mark_lines('abs', z=lens_z[i], usesmooth=True, markatm=False)
    plt.figure(figsize=(6, 3))
    plt.plot(ssp['wav'], ssp['flux']/np.sqrt(ssp['var']))
    median_snr.append(np.median(ssp['flux']/np.sqrt(ssp['var'])))
    mean_snr.append(np.mean(ssp['flux']/np.sqrt(ssp['var'])))
    print('median SNR: %f mean SNR: %f' %(median_snr[i], mean_snr[i]))
```

```python
print(mean_snr)
print(median_snr)
```
