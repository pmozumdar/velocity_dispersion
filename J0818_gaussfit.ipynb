```python
from veldis.gaussian_fit import Gaussfit
import numpy as np
from astropy.io import fits as pf
import matplotlib.pyplot as plt
```

```python
spec = Gaussfit('../spectra_extraction/J0818_ESI_spectra/spec1d_J0818_coadd2d_deflector.fits',
                informat='fits', logwav=True, trimsec=[4200,-100])
```

```python
%matplotlib notebook
spec.plot()
```

```python
dt = pf.open('../galaxy_spectra/ESI_data/2019_04/J0818-2613_0038_crsub.fits')
```

```python
dt[0].data.shape
```

```python
sp = np.nanmedian(dt[0].data[1730:, :], axis=0)
```

```python
plt.plot(np.arange(len(sp)), sp)
```

```python
spec.sky = sp
spec.wav = np.arange(len(sp))
```

```python
spec.plot()
```

```python
sky_lines = [(246, 266), (341, 359.5), (448, 465), (897, 914), (950.7, 965.6), (1330, 1350),
             (1389, 1407), (1681, 1711), (1723, 1738), (1971, 1989), (2095, 2112), (2232, 2248), 
             (2612, 2629), (2726, 2752), (2858, 2885)]
```

```python
best_fit = spec.fit_gauss(sky_lines)
```

```python
spec.sigma_inst(best_fit=best_fit, ylim=(1.8, 2.8))
```
